```python
import pandas as pd
```

**Subir a base para fazermos as recomendações**

**Adicional**

Processo essencial que fornecerá informações valiosas ao cliente, com sugestões de ações para reduzir o consumo de energia e realizar manutenções preventivas.

```python
df_consumo = pd.read_csv('base_cliente')
df_consumo
```

**Código para agrupar por mês e calcular a média de consumo diário:**

```python
df_consumo['data'] = pd.to_datetime(df_consumo['data'])

df_consumo['ano_mes'] = df_consumo['data'].dt.to_period('M')
```

```python
df_consumo
```

```python
# Calcular a média de consumo diário por mês
consumo_mensal = df_consumo.groupby('ano_mes').agg(
    media_consumo_diario=('consumo_kwh', 'mean')
).reset_index()
```

```python
consumo_mensal
```

```python
media_consumo_diario = consumo_mensal['media_consumo_diario'].mean()
media_consumo_diario
```

**Identificar o item que gastou mais por mês.**

```python
# Agrupar por mês e identificar o item com maior consumo
df = df_consumo
item_mais_consumo = df.groupby(['ano_mes', 'item'])['consumo_kwh'].sum().reset_index()

# Identificar o item com maior consumo por mês
item_mais_consumo_mensal = item_mais_consumo.loc[item_mais_consumo.groupby('ano_mes')['consumo_kwh'].idxmax()]

# Mostrar o item que mais consumiu por mês
print(item_mais_consumo_mensal[['ano_mes', 'item', 'consumo_kwh']])
```

**Identificar o cômodo que gastou mais por mês.**

```python
# Agrupar por mês e identificar o cômodo com maior consumo
comodo_mais_consumo = df.groupby(['ano_mes', 'ambiente'])['consumo_kwh'].sum().reset_index()

# Identificar o cômodo com maior consumo por mês
comodo_mais_consumo_mensal = comodo_mais_consumo.loc[comodo_mais_consumo.groupby('ano_mes')['consumo_kwh'].idxmax()]

# Mostrar o cômodo que mais consumiu por mês
print(comodo_mais_consumo_mensal[['ano_mes', 'ambiente', 'consumo_kwh']])
```

**Calcular a variação de consumo**

```python
# Agrupar o consumo por ano e mês, somando o consumo de cada mês
df_mensal = df.groupby('ano_mes')['consumo_kwh'].sum().reset_index()

# Calcular a variação de consumo entre os meses (percentual)
df_mensal['variacao_mes'] = df_mensal['consumo_kwh'].pct_change() * 100 

# Exibir os resultados
print(df_mensal[['ano_mes', 'consumo_kwh', 'variacao_mes']])
```

**Criar o Relatório**

```python
# Unir as tabelas df_mensal, item_mais_consumo_mensal e comodo_mais_consumo_mensal com base na coluna 'ano_mes'
df_relatorio = df_mensal.merge(item_mais_consumo_mensal, on='ano_mes', how='left')
df_relatorio = df_relatorio.merge(comodo_mais_consumo_mensal, on='ano_mes', how='left')
df_relatorio = df_relatorio.merge(consumo_mensal, on='ano_mes', how='left')

# Exibir uma amostra do dataframe unido
df_relatorio.head()
```

```python
def gerar_relatorio(df_relatorio):
    relatorios = []
    
    # Para cada mês, gerar um relatório
    for index, row in df_relatorio.iterrows():
        relatorio = f"""
        Relatório de Consumo - {row['ano_mes']}
        
        1. Consumo médio diário: {row['media_consumo_diario']:.2f} kWh/dia
        2. Item que mais gastou: {row['item']} com {row['consumo_kwh_x']:.2f} kWh
        3. Cômodo que mais gastou: {row['ambiente']} com {row['consumo_kwh_y']:.2f} kWh
        4. Variação do consumo em relação ao mês anterior: {row['variacao_mes']:.2f}%
        
        """
        relatorios.append(relatorio)
    
    # Juntar todos os relatórios
    return "\n".join(relatorios)

# Gerar o relatório completo
relatorio_completo = gerar_relatorio(df_relatorio)

# Exibir o relatório
print(relatorio_completo)
```

**Criar sugestão**

```python
def gerar_sugestao(row):
    # Se a variação de consumo for positiva
    if row['variacao_mes'] > 0:
        if row['item'] == 'luz':
            return f"Sugestão: O consumo de {row['item']} aumentou. Considere substituir lâmpadas por LED e verificar a fiação para evitar desperdícios. Se necessário, consulte um eletricista."
        elif row['item'] == 'aquecedor':
            return f"Sugestão: O consumo do {row['item']} aumentou. Verifique a eficiência do aquecedor e considere a manutenção ou a troca por um modelo mais econômico."
        elif row['item'] == 'ar condicionado':
            return f"Sugestão: O consumo do {row['item']} aumentou. Verifique a manutenção do aparelho ou considere substituí-lo por um modelo mais eficiente."
        # Para outros itens, podemos adicionar sugestões personalizadas também
        else:
            return f"Sugestão: O consumo do item {row['item']} aumentou. Verifique a eficiência do aparelho e considere a manutenção ou substituição por um modelo mais eficiente."

    # Se a variação de consumo for negativa
    elif row['variacao_mes'] < 0:
        if row['item'] == 'luz':
            return f"Sugestão: O consumo de {row['item']} diminuiu. Parabéns pela economia! Continue monitorando para manter a redução."
        elif row['item'] == 'aquecedor':
            return f"Sugestão: O consumo do {row['item']} diminuiu. Continue utilizando o equipamento de forma eficiente para manter os custos baixos."
        elif row['item'] == 'ar condicionado':
            return f"Sugestão: O consumo do {row['item']} diminuiu. Ótimo! Continue monitorando o uso para evitar desperdícios."
        # Para outros itens, podemos criar sugestões similares
        else:
            return f"Sugestão: O consumo do item {row['item']} diminuiu. Continue com o uso consciente dos aparelhos."

    # Caso o consumo permaneça estável (sem variação)
    else:
        return f"Sugestão: O consumo se manteve estável. Continue monitorando para garantir que os custos permaneçam controlados."

# Aplicar a função para gerar a coluna de sugestões
df_relatorio['sugestao'] = df_relatorio.apply(gerar_sugestao, axis=1)

# Exibir as primeiras linhas para verificar
print(df_relatorio[['ano_mes', 'item', 'consumo_kwh_x', 'variacao_mes', 'ambiente', 'consumo_kwh_y', 'sugestao']].head())

df_relatorio.to_excel('relatorio_sugestao.xlsx')
```
